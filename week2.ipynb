```python
import numpy as np

def generate_kernel(dim, n_points):
    points = np.random.rand(dim, n_points)
    X = points.T @ points
    return X
```

```python
def estimate_C(m):
    X = generate_kernel(2, 100)
    A = X[:m, :m]
    B = X[:m, m:]
    C = X[m:, m:]
    C_tilda = B.T @ np.linalg.pinv(A) @ B
    eps = np.linalg.norm(C - C_tilda, ord='fro')
    return eps
```

```python
import matplotlib.pyplot as plt


def plot_nystrom_error(m):
    results = [estimate_C(m) for _ in range(100)]

    plt.plot(results)
    plt.title(f'Nystrom Error for m={m}')
    plt.xlabel('Iteration')
    plt.ylabel('Error')
    plt.show()
```

```python
plot_nystrom_error(5)
plot_nystrom_error(4)
plot_nystrom_error(3)
plot_nystrom_error(2)
```
